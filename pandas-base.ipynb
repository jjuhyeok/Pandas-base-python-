# 1 : Series

```python
ob = pd.Series([4,7,-5,3]) #시리즈 정의하기
ob
```

```python
#Series 값 확인하기
ob.values
```

```python
#Series 인덱스 확인하기
ob.index
```

```python
#Series 자료형 확인하기
ob.dtype
```

```python
# 인덱스를 설정하는 법
ob2 = pd.Series([10,20,30,40], index = ['a','b','c','d'])
ob2
```

```python
#딕셔너리를 Seires data로 만들기
dic = {'chicken': 20000, 'pizza': 25000, 'rice': 1000, 'coke': 2000}
srdic = pd.Series(dic)
srdic
```

```python
#srdic의 이름 설정, 인덱스 이름 설정
srdic.name = 'Food'
srdic.index.name = 'Kind'
srdic
```

```python
#srdic의 인덱스 변경
srdic.index = ['치킨', '피자', '밥', '콜라']
srdic
```
